```python
import gwpy
from gwpy.timeseries import TimeSeries
strainData= TimeSeries.read('challenge1.gwf', 'H1:CHALLENGE1')
strainData.sample_rate
```

```python
import pylab
pylab.plot(strainData)
pylab.xlabel('Time(s)')
pylab.show()
```

```python
qt = strainData.q_transform(frange=(30,500), outseg=(60,75))
plot=qt.plot()
ax=plot.gca()
ax.set_yscale('log')
ax.colorbar(label='normalised energy')
```
